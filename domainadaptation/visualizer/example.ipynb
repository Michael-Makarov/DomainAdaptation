```python
import numpy as np
from sklearn.datasets import load_iris, make_classification, make_blobs, make_circles

# X_source, y_source = load_iris(return_X_y=True)
# X_source, y_source = make_classification(n_samples=1000, n_features=100, n_classes=12, n_informative=50)
X_source, y_source = make_blobs(n_samples=160, n_features=28, centers=12, cluster_std=3)

print('X_source shape: {}'.format(X_source.shape))
print('y_source shape: {}'.format(X_source.shape))
```

```python
X_target = X_source + 5
y_target = y_source.copy()

print('X_target shape: {}'.format(X_target.shape))
print('y_target shape: {}'.format(y_target.shape))
```

```python
X = np.concatenate([X_source, X_target], axis=0)
y = np.concatenate([y_source, y_target], axis=0)
domains = np.concatenate([
    ['source'] * len(X_source),
    ['target'] * len(X_target)
])

print('X shape: {}'.format(X.shape))
print('y shape: {}'.format(y.shape))
print('domains shape: {}'.format(domains.shape))
```

```python
from visualizer import Visualizer

visualizer = Visualizer(
    embeddings=X,
    domains=domains,
    labels=y,
    method='tsne')
```

```python
visualizer.visualize(size=75, filename='fig', draw_legend=False)
```
